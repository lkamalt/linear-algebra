### Задание 1 <a class='anchor' id='lin_space1_1'>
    
Исследуйте на линейную зависимость:

$$f_1(x) = e^x, \; f_2(x) = 1, \; f_3(x) = x + 1, \; f_4(x) = x - e^x.$$

**Решение**
  
Очевидно, что один из векторов является линейной комбинацией остальных:

$$f_1(x) = e^x = x + 1 - 1 - x + e^x = f_3(x) - f_2(x) - f_4(x),$$

из этого следует, что вектора *линейно зависимы*.
    
**Ответ (задание 1):** вектора линейно зависимы.

### Задание 2 <a class='anchor' id='lin_space1_2'>
    
Исследуйте на линейную зависимость:

$$f_1(x) = 2, \; f_2(x) = x, \; f_3(x) = x^2, \; f_4(x) = (x + 1)^2.$$

**Решение**
    
Очевидно, что один из векторов является линейной комбинацией остальных:
    
$$f_4(x) = (x + 1)^2 = x^2 + 2x + 1 = f_3(x) + 2 f_2(x) + \frac{1}{2} f_1(x),$$
    
из этого следует, что вектора *линейно зависимы*.
    
**Ответ (задание 2):** вектора линейно зависимы.

### Задание 3 <a class='anchor' id='lin_space1_3'>
    
Найдите координаты вектора $x = (2, 3, 5) \in \mathbb{R}^{3}$ в базисе $b_{1}=(0, 0, 10)$, $b_{2}=(2, 0, 0)$, $b_{3}=(0, 1, 0)$.

**Решение**
    
Обозначим новый базис буквой $B$. 
    
Вектор $x = (2, 3, 5) \in \mathbb{R}^{3}$ в новом базисе $B$ можно представить в виде:
  
$$x = (2, 3, 5) = (0, 0, 5) + (2, 0, 0) + (0, 3, 0) = \frac{1}{2} \cdot (0, 0, 10) + 1 \cdot (2, 0, 0) + 3 \cdot (0, 1, 0) = \frac{1}{2}b_1 + b_2 + 3b_3.$$

Таким образом, координаты вектора $x = (2, 3, 5)$ в базисе $B$: 
    
$$x' = \left(\frac{1}{2}, 1, 3 \right).$$
    
**Ответ (задание 3):** $x' = \left(\frac{1}{2}, 1, 3 \right)$.

### Задание 4 <a class='anchor' id='lin_space1_4'>
    
Найдите координаты вектора $3x^2 - 2x + 2 \in \mathbb{R}^3[x]$:

а) в базисе $1$, $x$, $x^2$;

б) в базисе $x^2$, $x - 1$, $1$.
    
**Решение**
    
а) В базисе $1$, $x$, $x^2$.
    
В заданном базисе вектор $3x^2 - 2x + 2$ можно представить в виде:
    
$$3x^2 - 2x + 2 = 2 \cdot 1 - 2 \cdot x + 3 \cdot x^2,$$
    
значит координаты вектора в этом базисе: $(2, -2, 3)$.
    
б) В базисе $x^2$, $x - 1$, $1$.
    
В заданном базисе вектор $3x^2 - 2x + 2$ можно представить в виде:
    
$$3x^2 - 2x + 2 = 3x^2 - 2(x - 1) = 3 \cdot x^2 - 2 \cdot (x - 1) + 0 \cdot 1,$$
    
значит координаты вектора в этом базисе: $(3, -2, 0)$.
    
**Ответ (задание 4):** а) $(2, -2, 3)$, б) $(3, -2, 0)$

### Задание 5 <a class='anchor' id='lin_space1_5'>
    
Установите, считается ли линейным подпространством:

а) совокупность всех векторов трёхмерного пространства, у которых по крайней мере одна из первых двух координат равна нулю;
    
б) все векторы, считающиеся линейными комбинациями данных векторов $\{u_1, u_2 \ldots, u_n\}$.

Для подпространств верно следующее утверждение:

Подмножество $L$ линейного пространства $V$ считается его подпространством только тогда, когда для любых элементов $u, v \in L$ и любого $\alpha \in \mathbb{R}$ выполняются условия:

1) $u + v \in L;$

2) $\alpha \cdot u \in L.$

Проверим эти условия для каждого варианта.

**Задание 5.а** <a class='anchor' id='lin_space1_5_1'>
    
Совокупность всех векторов трёхмерного пространства, у которых по крайней мере одна из первых двух координат равна нулю.
    
**Решение**
    
Рассматривается набор векторов $L$ следующего вида $(0, a, b)$, $(0, 0, a)$, $(d, 0, e)$.
    
Очевидно, что не выполняется первое условие. 
    
Для проверки возьмем вектора $u = (0, a, b)$ и $v = (d, 0, e)$, где $u, v \in L$:
    
$$u + v = (0, a, b) + (d, 0, e) = (d, a, b + e) \notin L,$$
    
то есть заданное множество *не является* линейным подпространством.

**Задание 5.б** <a class='anchor' id='lin_space1_5_2'>
    
Все векторы, считающиеся линейными комбинациями данных векторов $\{u_1, u_2 \ldots, u_n\}$.
    
**Решение**
    
Рассматривается набор векторов $L$ вида: 
    
$$\sum_{i = 1}^n \lambda_i u_i,$$
    
где $\lambda_i$ одновременно не равны $0$.
    
Очевидно, что оба условия выполняются. 
    
Для проверки возьмем вектора $u = \sum_{i = 1}^n \lambda_i u_i$ и $v = \sum_{i = 1}^n \alpha_i u_i$, где $u, v \in L$, тогда

$$u + v = \sum_{i = 1}^n \lambda_i u_i + \sum_{i = 1}^n \alpha_i u_i = \sum_{i = 1}^n (\lambda_i + \alpha_i) u_i \in L,$$
    
$$\alpha \cdot u = \alpha \sum_{i = 1}^n \lambda_i u_i = \sum_{i = 1}^n (\alpha \lambda_i) u_i \in L,$$
    
то есть заданное множество *является* линейным подпространством.

**Ответ (задание 5):** а) нет, б) да.

### Задание 1 <a class='anchor' id='lin_space2_1'>
    
Найдите скалярное произведение векторов $x, y \in \mathbb{R}$:
    
а) $x = (0, -3, 6), y = (-4, 7, 9);$
    
б) $x = (7, -4, 0, 1), y = (-3, 1, 11, 2).$
    
**Решение**

а) $x = (0, -3, 6), y = (-4, 7, 9)$.
    
Скалярное произведение:

$$(x, y) = \sum_i x_i y_i = 0 \cdot (-4) + (-3) \cdot 7 + 6 \cdot 9 = 33.$$

    
б) $x = (7, -4, 0, 1), y = (-3, 1, 11, 2)$.
    
Скалярное произведение:

$$(x, y) = \sum_i x_i y_i = 7 \cdot (-3) + (-4) \cdot 1 + 0 \cdot 11 + 1 \cdot 2 = -23.$$

```python
# Проверка через numpy
x = np.array([0, -3, 6])
y = np.array([-4, 7, 9])
dot_prod1 = np.dot(x, y)

x = np.array([7, -4, 0, 1])
y = np.array([-3, 1, 11, 2])
dot_prod2 = np.dot(x, y)

print(f'Проверка через numpy:\nа) (x, y) =  {dot_prod1}\nб) (x, y) = {dot_prod2}')
```

**Ответ (задание 1):** а) 33, б) -23.

### Задание 2 <a class='anchor' id='lin_space2_2'>
    
Найдите нормы векторов $x = (4, 2, 4)$ и $y = (12, 3, 4)$ и угол между ними.
   
**Решение**

В нормированном евклидовом пространстве косинус угла $\varphi$ между двумя между двумя произвольными элементами $x$ и $y$ вычисляется следующим образом:

$$\cos\varphi = \frac{(x,y)}{\left \|x\right \|\cdot\left \|y\right \|}.$$

Норма любого элемента $x$ в нормированном евклидовом пространстве определяется через скалярное произведение:

$$\left\|x\right\| = \sqrt{(x, x)}.$$

Скалярное произведение векторов $x$, $y$:

$$(x, y) = \sum_i x_i y_i = 4 \cdot 12 + 2 \cdot 3 + 4 \cdot 4 = 70.$$

Норма вектора $x$:

$$\left\|x\right\| = \sqrt{\sum_i x_i^2} = \sqrt{4^2 + 2^2 + 4^2} = \sqrt{36} = 6.$$

Норма вектора $y$:

$$\left\|y\right\| = \sqrt{\sum_i y_i^2} = \sqrt{12^2 + 3^2 + 4^2} = \sqrt{169} = 13.$$

Тогда косинус угла $\varphi$ между векторами $x$ и $y$ определяется следующим образом:

$$\cos\varphi = \frac{70}{6 \cdot 13} \approx 0.897.$$

и угол

$$\varphi = \arccos \left(\frac{70}{6 \cdot 13} \right) \approx 0.457 \; \text{рад} \approx 26.177 \; \text{град}.$$

```python
# Расчет
xy_dot_prod = 4 * 12 + 2 * 3 + 4 * 4
x_norm = np.sqrt(4 * 4 + 2 * 2 + 4 * 4)
y_norm = np.sqrt(12 * 12 + 3 * 3 + 4 * 4)
cos_phi = xy_dot_prod / (x_norm * y_norm)

print('Ручной расчет:')
print(f'(x, y) = {xy_dot_prod}\nx_norm = {x_norm}, y_norm = {y_norm}')
print(f'cos_phi = {cos_phi}\n')

# Проверка через numpy
x = np.array([4, 2, 4])
y = np.array([12, 3, 4])

xy_dot_prod = np.dot(x, y)
x_norm = np.linalg.norm(x)
y_norm = np.linalg.norm(y)

cos_phi = xy_dot_prod / (x_norm * y_norm)
phi = np.arccos(cos_phi)

print('Проверка через numpy:')
print(f'(x, y) = {xy_dot_prod}\nx_norm = {x_norm}, y_norm = {y_norm}')
print(f'cos_phi = {cos_phi}\nphi = {phi} рад\nphi = {np.rad2deg(phi)} град')
```

**Ответ (задание 2):** $\varphi \approx 26.177 \; \text{град}$.

### Задание 3 <a class='anchor' id='lin_space2_3'>
    
Определите, будет ли линейное пространство евклидовым, если за скалярное произведение принять:
    
а) произведение длин векторов;
    
б) утроенное обычное скалярное произведение векторов.

Аксиомы для скалярного произведения векторов $x$, $y$ в евклидовом пространстве:

1) $(x, y) = (y, x).$

2) $(\lambda x, y) = \lambda(x, y).$

3) $(x_1 + x_2, y) = (x_1, y) + (x_2, y).$

4) $(x, x)\ge 0$, причём $(x, x) = 0 \Leftrightarrow x=0.$

В данном примере проверяемое скалярное произведение будет обозначаться угловыми скобками:

$$\langle x, y \rangle,$$

а обычное - круглыми скобками:

$$(x, y),$$

для того, чтобы различать обычное и проверяемое скалярное произведение. 

**Задание 3.а** <a class='anchor' id='lin_space2_3_1'>
    
Скалярное произведение двух векторов = произведение длин этих векторов.
    
**Решение**
    
Утверждается, что 

$$\langle x, y \rangle = \left\|x\right\| \cdot \left\|y\right\|.$$
    
Проверим аксиомы скалярного произведения.
    
1) $\langle x, y \rangle = \left\|x\right\| \cdot \left\|y\right\| = \left\|y\right\| \cdot \left\|x\right\| = \langle y, x \rangle$ - выполняется.
    
2) $\langle \lambda x, y \rangle = \left\|\lambda x\right\| \cdot \left\|y\right\| = |\lambda| \left\|x\right\| \cdot \left\|y\right\| = |\lambda| \langle x, y \rangle \neq \lambda \langle x, y \rangle$ - не выполняется.
    
3) $\langle x_{1} + x_{2}, y \rangle = \left\|x_1 + x_2\right\| \cdot \left\|y\right\| \le \left( \left\|x_1\right\| + \left\|x_2\right\| \right) \left\|y\right\| = \left\|x_1\right\| \cdot \left\|y\right\| + \left\|x_2\right\| \cdot \left\|y\right\| = \langle x_1, y \rangle + \langle x_2, y \rangle$ - не выполняется.
    
4) $\langle x, x \rangle = \left\|x\right\| \cdot \left\|x\right\| = \left\|x\right\|^2 \ge 0$.

Очевидно, что величина $\langle x, x \rangle = \left\|x\right\|^2$ всегда больше или равна нулю, причем она равна нулю тогда и только тогда, когда $x$ - нулевой вектор, то есть четвертая аксиома выполняется.
 
Таким образом, не выполняются две аксиомы, значит линейное пространство с заданным скалярным произведением *не является* евклидовым.

**Задание 3.б** <a class='anchor' id='lin_space2_3_2'>
    
Скалярное произведение двух векторов = утроенное обычное скалярное произведение векторов.
    
**Решение**
    
Утверждается, что 

$$\langle x, y \rangle = 3(x, y).$$
    
Проверим аксиомы скалярного произведения.
    
1) $\langle x, y \rangle = 3(x, y) = 3(y, x) = \langle y, x \rangle$ - выполняется.
    
2) $\langle \lambda x, y \rangle = 3(\lambda x, y) = 3\lambda(x, y) = \lambda \langle x, y \rangle$ - выполняется.
    
3) $\langle x_1 + x_2, y \rangle = 3(x_1 + x_2, y) = 3(x_1, y) + 3(x_2, y) = \langle x_1, y \rangle + \langle x_2, y \rangle$ - выполняется.
    
4) $\langle x, x \rangle = 3(x, x) \ge 0$.

По свойствам скалярного произведения величина $\langle x, x \rangle = 3(x, x)$ равна нулю, тогда и только тогда, когда $x = 0$, значит четвертая аксиома тоже выполняется.
    
Таким образом, линейное пространство с заданным скаларным произведением *является* евклидовым.

**Ответ (задание 3):** а) нет, б) да.

### Задание 4 <a class='anchor' id='lin_space2_4'>
    
Выясните, какие из нижеперечисленных векторов образуют ортонормированный базис в линейном пространстве $\mathbb{R}^3$:
    
а) $(1, 0, 0), \; (0, 0, 1);$
    
б) $\left(\frac{1}{\sqrt{2}}, -\frac{1}{\sqrt{2}}, 0\right), \; \left(\frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}}, 0\right), \; (0, 0, 1);$
    
в) $\left(\frac{1}{2}, -\frac{1}{2}, 0 \right), \; \left(0, \frac{1}{2}, \frac{1}{2} \right), \; (0, 0, 1);$
    
г) $(1, 0, 0), \; (0, 1, 0), \; (0, 0, 1).$ 

В конечномерном евклидовом пространстве базис $\{e_{1},e_{2}...,e_{n}\}$ называется ортонормированным, если:

1) $(e_{i}, e_{j})=0$ $\forall$ $i\neq j$ - вектора ортогональны друг другу;

2) $(e_{i},e_{i})=1$ $\forall$ $i\in[1, n]$ - все вектора единичной длины.

Для каждого варианта набора векторов будут проверены эти свойства.

**Задание 4.а** <a class='anchor' id='lin_space2_4_1'>
    
$(1, 0, 0), \; (0, 0, 1)$

**Решение**

Очевидно, что в линейном пространстве $\mathbb{R}^3$ набор из двух векторов не может образовывать базис, потому что по определению линейное пространство является $n$-мерным, если в нем есть $n$ линейно независимых векторов, называемых базисом.

Таким образом, в данном случае должно быть три вектора.

Однако проверим указанные свойства и для них тоже.

Первое свойство:

$$(e_1, e_2) = 0,$$

Певрое свойство выполняется.

Второе свойство:

$$(e_1, e_1) = 1 + 0 + 0 = 1,$$

$$(e_2, e_2) = 0 + 0 + 1 = 1,$$

второе свойство тоже выполняется.

В целом свойства выполняются, но в силу определения $n$-мерного пространства, такой набор из двух векторов *не может образовывать* базис в $\mathbb{R}^3$.

**Задание 4.б** <a class='anchor' id='lin_space2_4_2'>

видно, что не выполняется первое свойство, значит можно сделать вывод о том, что заданные вектора *не образуют* ортонормированный базис в линейном пространстве $\mathbb{R}^3$.

**Задание 4.г** <a class='anchor' id='lin_space2_4_4'>
    
$e_1 = (1, 0, 0), \; e_2 = (0, 1, 0), \; e_3 = (0, 0, 1)$
    
**Решение**

$$(e_1, e_2) = 1 \cdot 0 + 0 \cdot 1 + 0 \cdot 0 = 0,$$

$$(e_1, e_3) = 1 \cdot 0 + 0 \cdot 0 + 0 \cdot 1 = 0,$$

$$(e_2, e_3) = 0 \cdot 0 + 1 \cdot 0 + 0 \cdot 1 = 0.$$

Певрое свойство выполняется.

Проверим второе свойство:

$$(e_2, e_2) = 1 + 0 + 0 = 1,$$

Второе свойство тоже выполняется, то есть действительно указанные вектора *образуют* ортонормированный базис в $\mathbb{R}^3$.
