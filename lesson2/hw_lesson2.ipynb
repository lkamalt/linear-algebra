### Задание 1 <a class='anchor' id='matrix1_1'>
    
Установите, какие произведения матриц $AB$ и $BA$ определены, и найдите размерности полученных матриц:

а) $A$ — матрица $4 \times 2$, $B$ — матрица $4 \times 2$;

б) $A$ — матрица $2 \times 5$, $B$ — матрица $5 \times 3$;

в) $A$ — матрица $8 \times 3$, $B$ — матрица $3 \times 8$;

г) $A$ — квадратная матрица $4 \times 4$, $B$ — квадратная матрица $4 \times 4$.
    
**Решение**

Для того, чтобы произведение матриц было определено, необходимо, чтобы число столбцов первой матрицы совпадало с числом строк второй матрицы.
 
а) $A$ — матрица $4 \times 2$, $B$ — матрица $4 \times 2$.

Число столбцов матрицы $A = 2$, число строк в матрице $B = 4$, значит умножение матриц $AB$ *не определено*.
    
Число столбцов матрицы $B = 2$, число строк в матрице $A = 4$, значит умножение матриц $BA$ *не определено*.

б) $A$ — матрица $2 \times 5$, $B$ — матрица $5 \times 3$.

Число столбцов матрицы $A = 5$, число строк в матрице $B = 5$, значит умножение матриц $AB$ *определено*.
    
Результирующая матрица $C = AB$ будет иметь размерность $2 \times 3$.
    
Число столбцов матрицы $B = 3$, число строк в матрице $A = 2$, значит умножение матриц $BA$ *не определено*.
    
в) $A$ — матрица $8 \times 3$, $B$ — матрица $3 \times 8$.

Число столбцов матрицы $A = 3$, число строк в матрице $B = 3$, значит умножение матриц $AB$ *определено*.
    
Результирующая матрица $C = AB$ будет иметь размерность $8 \times 8$.
    
Число столбцов матрицы $B = 8$, число строк в матрице $A = 8$, значит умножение матриц $BA$ *определено*.
    
Результирующая матрица $C = BA$ будет иметь размерность $3 \times 3$.
    
г) $A$ — квадратная матрица $4 \times 4$, $B$ — квадратная матрица $4 \times 4$.

Матрицы $A$ и $B$ являются квадратными размерности $4 \times 4$, значит произведения матриц $AB$ и $BA$ *определены*.
    
Результирующие матрицы $C_1 = AB$ и $C_2 = BA$ будут иметь размерность $4 \times 4$.
 
**Ответ (задание 1):** <br>
а) $AB$ - нет, $BA$ - нет, <br>
б) $AB$ - да $2 \times 3$, $BA$ - нет; <br>
в) $AB$ - да $8 \times 8$, $BA$ - да $3 \times 3$; <br>
г) $AB$ - да $4 \times 4$, $BA$ - да $4 \times 4$.

### Задание 2 <a class='anchor' id='matrix1_2'>
    
Найдите сумму и произведение матриц 
$A = \begin{pmatrix}
1 & -2 \\ 
3 & 0
\end{pmatrix}$ и 
$B = \begin{pmatrix}
4 & -1 \\ 
0 & 5
\end{pmatrix}.$
    
**Решение**
 
Сумма матриц:

$$
A + B = 
\begin{pmatrix}
1 & -2 \\ 
3 & 0
\end{pmatrix} + 
\begin{pmatrix}
4 & -1 \\ 
0 & 5
\end{pmatrix} = 
\begin{pmatrix}
1 + 4 & -2 - 1 \\ 
3 + 0 & 0 + 5
\end{pmatrix} =
\begin{pmatrix}
5 & -3 \\ 
3 & 5
\end{pmatrix}.
$$
    
Произведение матриц:
    
$$
A \cdot B = 
\begin{pmatrix}
1 & -2 \\ 
3 & 0
\end{pmatrix} \cdot 
\begin{pmatrix}
4 & -1 \\ 
0 & 5
\end{pmatrix} = 
\begin{pmatrix}
1 \cdot 4 + (-2) \cdot 0 & 1 \cdot (-1) + (-2) \cdot 5  \\ 
3 \cdot 4 + 0 \cdot 0 & 3 \cdot (-1) + 0 \cdot 5
\end{pmatrix} =
\begin{pmatrix}
4 & -11 \\ 
12 & -3
\end{pmatrix}.
$$

```python
# Проверка через numpy
A = np.array([[1, -2], [3, 0]])
B = np.array([[4, -1], [0, 5]])

sum_AB = A + B
dot_AB = np.dot(A, B)

print(f'A + B = \n{sum_AB}')
print(f'A * B = \n{dot_AB}')
```

**Ответ (задание 2):** 
$
A + B = 
\begin{pmatrix}

3 & 5
\end{pmatrix}$,

### Задание 5 <a class='anchor' id='matrix1_5'>
    
Напишите на Python функцию для перемножения двух произвольных матриц, не используя NumPy.

```python
from itertools import chain
from operator import add, sub


class Order:
    """ Тип обхода по матрице: C - row-major, F - column-major """
    C, F = range(2)


class Matrix:
    """ Класс для описания объекта матрицы """
    def __init__(self, list_of_rows):
        """
        Конструктор класса
        :param list_of_rows: данные матрицы
        :type list_of_rows: List[List[float]]
        """
        if not list_of_rows:
            raise ValueError('Пустой список')
        if len(set(map(len, list_of_rows))) != 1:
            raise ValueError('Строки разной размерности')

        self.data = list_of_rows
        self.rows = len(list_of_rows)
        self.cols = len(list_of_rows[0]) if list_of_rows else 0

    def __add__(self, other):
        if not self.same_dims(other):
            raise ValueError('Разные размерности')
        data_sum = [list(map(add, self.data[i], other.data[i])) for i in range(self.rows)]
        return Matrix(data_sum)

    def __sub__(self, other):
        if not self.same_dims(other):
            raise ValueError('Разные размерности')
        data_sum = [list(map(sub, self.data[i], other.data[i])) for i in range(self.rows)]
        return Matrix(data_sum)

    def __mul__(self, other):
        return self.dot(other)

    def __str__(self):
        return '\n'.join(['\t'.join(map(str, row)) for row in self.data])

    def same_dims(self, other):
        """
        Определяет равны или нет размерности матрицы
        :param other: матрица, с которой происходит сравнение
        :type other: Matrix
        :rtype: bool
        """
        return self.rows == other.rows and self.cols == other.cols

    def get_flatten(self, order=Order.C):
        """
        Возвращает список из элементов матрицы в заданном порядке
        Если Order.C - обход по строкам, если Order.F - по столбцам
        :param order: тип порядка обхода по матрице
        :type order: Order
        :rtype: List[float]
        """
        if order == Order.F:
            return list(chain(*self.data))
        return list(chain(*zip(*self.data)))

    def dot(self, other):
        """
        Возвращает произведение матриц
        :param other: матрица, на которую умножается
        :type other: Matrix
        :rtype: Matrix
        """
        if self.cols != other.rows:
            raise ValueError('Количество столбцов первой матрицы не совпадает с количеством строк второй матрицы')
        data_mul = [[None for _ in range(other.cols)] for _ in range(self.rows)]
        for i in range(self.rows):
            for j in range(other.cols):
                data_mul[i][j] = sum(self.data[i][p] * other.data[p][j] for p in range(self.cols))
        return Matrix(data_mul)

    def transpose(self):
        """
        Возвращает транспонированную матрицу
        :rtype: Matrix
        """
        data = list(map(list, zip(*self.data)))
        return Matrix(data)
```

```python
# Проверка
A = Matrix([[4, 1], [5, -2], [2, 3]])
A_T = A.transpose()
print(f'A =\n{A}\n')
print(f'A^T =\n{A_T}\n')
print(f'A * A^T =\n{A * A.transpose()}\n')
print(f'A * A^T =\n{A.dot(A_T)}')
```

### Задание 3 <a class='anchor' id='matrix2_3'>
    
Докажите, что матрица:

$$\begin{pmatrix}
-2 & 7 & -3 \\ 
4 & -14 & 6 \\ 
-3 & 7 & 13
\end{pmatrix}$$
   
вырожденная.
    
**Решение**

Матрица является вырожденной, если её определитель равен $0$.
    
Очевидно, что в этой матрице первая и вторая строка - линейно зависимы, так как вторую строку можно получить из первой, умножив первую на $-2$. 
    
Тогда огласно свойствам определитель этой матрицы равен $0$, значит матрица *является* вырожденной. 

Чтобы проверить, вычислим определитель заданной матрицы:
    
$$
\begin{vmatrix}
-2 & 7 & -3 \\ 
4 & -14 & 6 \\ 
-3 & 7 & 13
\end{vmatrix} = 
-2 \begin{vmatrix}
-14 & 6 \\ 
7 & 13 
\end{vmatrix} - 
7 \begin{vmatrix}
4 & 6 \\ 
-3 & 13 
\end{vmatrix} -
3 \begin{vmatrix}
4 & -14 \\ 
-3 & 7 
\end{vmatrix} = -2(-182 - 42) - 7(52 + 18) - 3(28 - 42) = 
$$
$$ = 2 \cdot 224 - 7 \cdot 70 + 3 \cdot 14 = 448 - 490 + 42 = 0.$$
    


```python
# Проверка через numpy
A = np.array([[-2, 7, -3], [4, -14, 6], [-3, 7, 13]])
print(f'|A| = {np.linalg.det(A)}')
```

**Ответ (задание 3):** матрица является вырожденной, потому что её определитель равен $0$.

### Задание 4 <a class='anchor' id='matrix2_4'>
    
Найдите ранг матрицы.
    
**Решение**
    
а) Определим ранг матрицы
$$
\begin{pmatrix}
1 & 2 & 3 \\ 
1 & 1 & 1 \\ 
2 & 3 & 4 \\
\end{pmatrix}.
$$

Вычтем из третьей строки первую, умноженную на 2:

$$
\begin{pmatrix}
1 & 2 & 3 \\ 
1 & 1 & 1 \\ 
0 & -1 & -2 \\
\end{pmatrix}.
$$
    
Вычтем из второй строки первую:

$$
\begin{pmatrix}
1 & 2 & 3 \\ 
0 & -1 & -2 \\ 
0 & -1 & -2 \\
\end{pmatrix}.
$$
    
Видно, что третья строка является линейной комбинацией первых двух, значит её можно отбросить:

$$
\begin{pmatrix}
1 & 2 & 3 \\ 
0 & -1 & -2 \\ 
\end{pmatrix}.
$$
    
Таким образом, ранг матрицы равен 2.

б) Определим ранг матрицы  
    
$$
\begin{pmatrix}
0 & 0 & 2 & 1 \\ 
0 & 0 & 2 & 2 \\ 
0 & 0 & 4 & 3 \\ 
2 & 3 & 5 & 6
\end{pmatrix}.
$$
    
Переставим первую и четвертую строку местами:
    
$$
\begin{pmatrix}
2 & 3 & 5 & 6 \\
0 & 0 & 2 & 2 \\ 
0 & 0 & 4 & 3 \\ 
0 & 0 & 2 & 1 \\ 
\end{pmatrix}.
$$ 

Вычтем из четвертой строки вторую, а из третьей - вторую, умноженную на 2:
    
$$
\begin{pmatrix}
2 & 3 & 5 & 6 \\
0 & 0 & 2 & 2 \\ 
0 & 0 & 0 & -1 \\ 
0 & 0 & 0 & -1 \\ 
\end{pmatrix}.
$$ 
    
Вычтем из четвертой строки третью:
    
$$
\begin{pmatrix}
2 & 3 & 5 & 6 \\
0 & 0 & 2 & 2 \\ 
0 & 0 & 0 & -1 \\ 
0 & 0 & 0 & 0 \\ 
\end{pmatrix}.
$$
    
Отбросим полученную строку из нулей:
    
$$
\begin{pmatrix}
2 & 3 & 5 & 6 \\
0 & 0 & 2 & 2 \\ 
0 & 0 & 0 & -1 \\ 
\end{pmatrix}.
$$ 
    
Таким образом, ранг матрицы равен 3.

```python
# Проверка через numpy
A_1 = np.array([[1, 2, 3], [1, 1, 1], [2, 3, 4]])
A_2 = np.array([[0, 0, 2, 1], [0, 0, 2, 2], [0, 0, 4, 3], [2, 3, 5, 6]])

print(f'rank_A_а = {np.linalg.matrix_rank(A_1)}')
print(f'rank_A_б = {np.linalg.matrix_rank(A_2)}')
```

**Ответ (задание 4):** <br>
а) $2$ <br>
б) $3$ <br>
