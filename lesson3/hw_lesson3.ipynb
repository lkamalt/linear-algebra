### Задание 1 <a class='anchor' id='lin_transf_1'>
    
Найдите собственные векторы и собственные значения для линейного оператора, заданного матрицей:

$$A = \begin{pmatrix}
-1 & -6 \\ 
2 & 6
\end{pmatrix}.$$
    
**Решение**
    
Чтобы число $\lambda$ было собственным значением оператора $\textbf{A}$, это число должно быть корнем характеристического уравнения оператора $\textbf{A}$.
    
Характеристическое уравнение матрицы $A$:
    
$$\text{det}(A - \lambda I) = \begin{vmatrix}
-1 - \lambda & -6 \\ 
2 & 6 - \lambda
\end{vmatrix} = -(1 + \lambda)(6 - \lambda) + 12 = -(6 + 6\lambda - \lambda - \lambda^2) + 12 = \lambda^2 - 5\lambda + 6 = 0.$$
    
Решим квдратное уравнение:
    
$$\lambda^2 - 5\lambda + 6 = 0,$$
    
$$D = b^2 - 4ac = 25 - 24 = 1,$$
    
$$\lambda_{1, 2} = \frac{-b \pm \sqrt{D}}{2a} = \frac{5 \pm 1}{2}.$$
    
Значит корни характеристического уравнения:
    
$$\lambda_1 = 3,$$
    
$$\lambda_2 = 2$$
    
являются собственными значениями матрицы $A$.
    
Найдем собственные векторы, подставив полученные собственные значения в выражение $(A - \lambda I)x = 0$.
    
1. $\lambda_1 = 3.$
    
$$(A - \lambda I)x = \begin{pmatrix}
-4 & -6 \\ 
2 & 3
\end{pmatrix}
\begin{pmatrix}
x_1 \\
x_2
\end{pmatrix} = 0.
$$
    
Распишем уравнения:
    
$$
\begin{cases}
-4 x_1 - 6 x_2 = 0, \\
2 x_1 + 3 x_2 = 0.
\end{cases}
\Rightarrow
\begin{cases}
2 x_1 + 3 x_2 = 0, \\
2 x_1 + 3 x_2 = 0,
\end{cases}
\Rightarrow
\begin{cases}
2 x_1 + 3 x_2 = 0,
\end{cases}
$$
    
видно, что вторая строка совпадает с первой, значит второе уравнение можно отбросить. 
    
Из полученного уравнения выразим $x_1$ через $x_2$:
    
$$x_1 = -\frac{3}{2}x_2.$$
    
Переменная $x_2$ может принимать любое значение, тогда приняв $x_2 = C_1$, получаем что, все вектора вида 

$$
\begin{pmatrix}
x_1 \\
x_2
\end{pmatrix} = \begin{pmatrix}
-\frac{3}{2}C_1 \\
C_1
\end{pmatrix}.
$$
    
где $C_1$ - произвольная константа, являются собственными векторами матрицы $A$.
    
2. $\lambda_2 = 2.$
    
$$(A - \lambda I)x = \begin{pmatrix}
-3 & -6 \\ 
2 & 4
\end{pmatrix}
\begin{pmatrix}
x_1 \\
x_2
\end{pmatrix} = 0.
$$
    
Распишем уравнения:
    
$$
\begin{cases}
-3 x_1 - 6 x_2 = 0, \\
2 x_1 + 4 x_2 = 0.
\end{cases}
\Rightarrow
\begin{cases}
x_1 + 2 x_2 = 0, \\
x_1 + 2 x_2 = 0,
\end{cases}
\Rightarrow
\begin{cases}
x_1 + 2 x_2 = 0,
\end{cases}
$$
    
видно, что и в этом случае вторая строка совпадает с первой, значит второе уравнение можно отбросить. 
    
Из полученного уравнения выразим $x_1$ через $x_2$:
    
$$x_1 = -2x_2.$$
    
Переменная $x_2$ может принимать любое значение, тогда приняв $x_2 = C_2$, получаем что, все вектора вида 

$$
\begin{pmatrix}
x_1 \\
x_2
\end{pmatrix} = \begin{pmatrix}
-2 C_2 \\
C_2
\end{pmatrix}.
$$
    
где $C_2$ - произвольная константа, являются собственными векторами матрицы $A$.
    
**Ответ (задание 1):**

Собственные значения: $\lambda_1 = 3$, $\lambda_2 = 2$
    
Собственные вектора: 
$\begin{pmatrix}
-\frac{3}{2}C_1 \\
C_1
\end{pmatrix}$,
$
\begin{pmatrix}
-2 C_2 \\
C_2
\end{pmatrix}
$, $C_1$, $C_2$ - произвольные константы

### Задание 2 <a class='anchor' id='lin_transf_2'>
    
Дан оператор поворота на 180 градусов, задаваемый матрицей:

$$A = \begin{pmatrix}
-1 & 0 \\ 
0 & -1
\end{pmatrix}.$$

Покажите, что __любой__ вектор считается для него собственным.
    
**Решение**
   
Возьмем некоторый вектор, сотоящий из произвольных констант:
    
$$x = \begin{pmatrix}
С_1 \\ 
С_2
\end{pmatrix},
$$
    
и предположим, что этот вектор $x$ является собственным вектором заданного линейного оператора, тогда должно существовать некоторое вещественное число $\lambda$, при котором $Ax = \lambda x$, то есть:

$$\begin{pmatrix}
-1 & 0 \\ 
0 & -1
\end{pmatrix}
\begin{pmatrix}
С_1 \\ 
С_2 
\end{pmatrix}=
\lambda
\begin{pmatrix}
С_1 \\ 
С_2
\end{pmatrix}.
$$

Из этого следует:

$$\begin{cases}
-С_1 = С_1 \cdot \lambda \\ 
-С_2 = С_2 \cdot \lambda
\end{cases}
\Rightarrow
\lambda = -1,$$
    
видно, что вектор $x = (С_1, С_2)$, состоящий из любых произвольных констант является собственным вектором заданного линейного оператора, причем его собственное значение $\lambda = -1$ для любого собственного вектора. Действительно оператор поворота на 180 градусов переводит любой вектор в противоположный по направлению вектор, сохраняя его длину.
    
**Ответ (задание 2):** любой вектор является собственным с собственным значением $\lambda = -1$

### Задание 3 <a class='anchor' id='lin_transf_3'>
    
Пусть линейный оператор задан матрицей:

$$A = \begin{pmatrix}
1 & 1 \\ 
-1 & 3
\end{pmatrix}.$$

Установите, считается ли вектор $x = (1, 1)$ собственным вектором этого линейного оператора.
    
**Решение**
 
Предположим, что вектор $x$ является собственным вектором заданного линейного оператора, тогда должно существовать некоторое вещественное число $\lambda$, при котором $Ax = \lambda x$, то есть:

$$\begin{pmatrix}
1 & 1 \\ 
-1 & 3
\end{pmatrix}
\begin{pmatrix}
1 \\ 
1 
\end{pmatrix}=
\lambda
\begin{pmatrix}
1 \\ 
1
\end{pmatrix}.
$$

Из этого следует:

$$\begin{cases}
1 + 1 = 1 \cdot \lambda \\ 
-1 + 3 = 1 \cdot \lambda
\end{cases}
\Rightarrow
\lambda = 2.$$

Таким образом, вектор $x = (1, 1)$ действительно является собственным вектором линейного оператора, заданного матрицей $A$, и его собственное значение $\lambda = 2$.

**Ответ (задание 3):** да, $\lambda = 2$

### Задание 4 <a class='anchor' id='lin_transf_4'>
    
Пусть линейный оператор задан матрицей:

$$A = \begin{pmatrix} 
0 & 3 & 0 \\ 
3 & 0 & 0 \\ 
0 & 0 & 3 
\end{pmatrix}.$$

Установите, считается ли вектор $x = (3, -3, -4)$ собственным вектором этого линейного оператора.
    
**Решение**
   
Предположим, что вектор $x$ является собственным вектором заданного линейного оператора, тогда должно существовать некоторое вещественное число $\lambda$, при котором $Ax = \lambda x$, то есть:

$$\begin{pmatrix}
0 & 3 & 0 \\ 
3 & 0 & 0 \\ 
0 & 0 & 3 
\end{pmatrix}
\begin{pmatrix}
3 \\ 
-3 \\
-4
\end{pmatrix}=
\lambda
\begin{pmatrix}
3 \\ 
-3 \\
-4
\end{pmatrix}.
$$

Из этого следует:

$$
\begin{cases}
-9 = 3 \cdot \lambda \\ 
9 = -3 \cdot \lambda \\
-12 = -4 \cdot \lambda
\end{cases}
\Rightarrow
\begin{cases}
\lambda = -3 \\ 
\lambda = -3 \\ 
\lambda = 3
\end{cases},
$$

видно, что такая система не имеет смысла, следовательно, вектор $x = (3, -3, -4)$ не является собственным вектором линейного оператора, заданного матрицей $A$.
    
**Ответ (задание 4):** нет
