### Задание 1 <a class='anchor' id='slau1_1'>
    
Решите систему уравнений методом Гаусса:

$$
\begin{cases}
x_1 + x_2 - x_3 - 2x_4 = 0, \\
2x_1 + x_2 - x_3 + x_4 = -2, \\
x_1 + x_2 - 3x_3 + x_4 = 4.
\end{cases}
$$

**Решение**
    
Расширенная матрица системы:
    
$$
\tilde{A} = \begin{pmatrix}
\left.\begin{matrix}
1 & 1 & -1 & -2 \\ 
2 & 1 & -1 & 1 \\ 
1 & 1 & -3 & 1
\end{matrix}\right|
\begin{matrix}
0 \\ 
-2 \\
4
\end{matrix}
\end{pmatrix}.
$$
    
Приведем матрицу к ступенчатому виду. 
    
Для этого сначала вычтем из второй строки первую, умноженную на 2:
    
$$
\tilde{A} = \begin{pmatrix}
\left.\begin{matrix}
1 & 1 & -1 & -2 \\ 
0 & -1 & 1 & 5 \\ 
1 & 1 & -3 & 1
\end{matrix}\right|
\begin{matrix}
0 \\ 
-2 \\
4
\end{matrix}
\end{pmatrix}.
$$
    
Далее прибавим первую строку к третьей:
    
$$
\tilde{A} = \begin{pmatrix}
\left.\begin{matrix}
1 & 1 & -1 & -2 \\ 
0 & -1 & 1 & 5 \\ 
0 & 0 & -2 & 3
\end{matrix}\right|
\begin{matrix}
0 \\ 
-2 \\
4
\end{matrix}
\end{pmatrix}.
$$
    
Из полученной матрицы видно, что система имеет бесконечное число решений. 
    
В качестве свободного параметра возьмем $x_4$, обозначим его через произвольную константу $C$:
    
$$x_4 = C.$$
    
Получим выражения для остальных неизвестных:
    
$$x_3 = -\frac{4 - 3x_4}{2} = \frac{3}{2}C - 2,$$
    
$$x_2 = 2 + 5x_4 + x_3 = 2 + 5c + \frac{3}{2}C - 2 = \frac{13}{2}C,$$
    
$$x_1 = 2x_4 + x_3 - x_2 = 2C + \frac{3}{2}C - 2 - \frac{13}{2}C = -3C - 2.$$
    
**Ответ (задание 1):** общее решение
$\begin{cases}
x_1 = -3C - 2 \\
x_2 = \frac{13}{2}C \\
x_3 = \frac{3}{2}C - 2 \\
x_4 = C
\end{cases}$, где $C$ - произвольная констната

### Задание 2 <a class='anchor' id='slau1_2'>
    
Проверьте на совместность и выясните, сколько решений будет иметь система линейных уравнений:

**Задание 2.а** <a class='anchor' id='slau1_2_1'>
    
$$
\begin{cases}
3x_1 - x_2 + x_3 = 4, \\
2x_1 - 5x_2 - 3x_3 = -17, \\
x_1 + x_2 - x_3 = 0;
\end{cases}
$$

**Решение**
    
Расширенная матрица системы:
        
$$
\tilde{A} = \begin{pmatrix}
\left.\begin{matrix}
3 & -1 & 1 \\
2 & -5 & -3 \\
1 & 1 & -1
\end{matrix}\right|
\begin{matrix}
4 \\ 
-17 \\
0
\end{matrix}
\end{pmatrix}.
$$
    
Очевидно, что столбцы исходной матрицы $A$ линейно независимы, а три линейно независимых вектора в трехмерном пространстве образуют базис, и любой другой вектор может быть представлен в виде линейной комбинации этих базисных векторов единственным образом. Поэтому эта система уравнений *совместна* и имеет *единственное решение*.
    
Докажем это утверждение с помощью теоремы Кронекера-Капелли, для этого определим ранги матриц.
    
Переставим строки матрицы:
    
$$
\tilde{A} = \begin{pmatrix}
\left.\begin{matrix}
1 & 1 & -1 \\

\end{matrix}\right|
\begin{matrix}
0 \\ 
-17 \\
4
\end{matrix}
\end{pmatrix}.
$$
    
Вычтем из второй строки первую, умноженную на 2:
    
$$
\tilde{A} = \begin{pmatrix}
\left.\begin{matrix}
1 & 1 & -1 \\
0 & -7 & -1 \\

\end{matrix}\right|
\begin{matrix}
0 \\ 
-17 \\
4
\end{matrix}
\end{pmatrix}.
$$
    
Вычтем из третьей строки первую, умноженную на 3:
    
$$
\tilde{A} = \begin{pmatrix}
\left.\begin{matrix}

\end{matrix}\right|
\begin{matrix}
0 \\

\end{matrix}
\end{pmatrix}.
$$
    
Разделим третью строку на 4 и поменяем местами со второй строокой:
    
$$
\tilde{A} = \begin{pmatrix}
\left.\begin{matrix}
1 & 1 & -1 \\
0 & -1 & 1 \\
0 & -7 & -1
\end{matrix}\right|
\begin{matrix}
0 \\ 
1 \\
-17
\end{matrix}
\end{pmatrix}.
$$
    
Вычтем из третьей строки вторую, умноженную на -7:
    
$$
\tilde{A} = \begin{pmatrix}
\left.\begin{matrix}
1 & 1 & -1 \\
0 & -1 & 1 \\
0 & 0 & -8
\end{matrix}\right|
\begin{matrix}
0 \\ 
1 \\
-24
\end{matrix}
\end{pmatrix}.
$$
    
Видно, что ранг исходной матрицы совпадает с рангом расширенной, причем ранг равен числу неизвестных:

$$\text{rank} A = \text{rank} \tilde A = 3,$$
    
значит по теореме Кронекера-Капелли система *совместна* и имеет *единственное решение*.

# Системы линейных уравнений. Часть 2 <a class='anchor' id='slau2'>

```python
def is_correct_LU(A, L, U):
    """ 
    Проверяет правильность LU-разложения и выводит результат в консоль
    :param A: исходная матрица
    :param L: матрица L, полученная в ходе разложения
    :param U: матрица U, полученная в ходе разложения
    """
    is_correct = np.isclose(A, L @ U).all()
    print(f'A == L @ U: {is_correct}')
```

```python
def show_solve_result(A, b):
    """ 
    Решает систему линейных уравнений и выводит результат в консоль
    :param A: матрица коэффициентов системы
    :param b: вектор правой части
    """
    print(f'Решение x = {np.linalg.solve(A, b)}')
```

$$A = \begin{pmatrix}
1 & -2 \\ 
3 & -4
\end{pmatrix}.$$
    
Вычислим определитель матрицы коэффициентов
    
$$
\text{det}A = \begin{vmatrix}
1 & -2 \\ 
3 & -4
\end{vmatrix} = -4 + 6 = 2.
$$
    
Определитель матрицы $\text{det}A = 2 \neq 0$ значит система совестна.

$$
\text{det}A_1 = \begin{vmatrix}

\end{vmatrix} = -4 + 14 = 10.
$$
    
Вычислим определитель матрицы, у которой второй столбец заменен правой частью:
    
$$
\text{det}A_2 = \begin{vmatrix}

3 & 7
\end{vmatrix} = 7 - 3 = 4.
$$
    
Найдем решение системы по формулам Крамера:
    
$$x_1 = \frac{\text{det}A_1}{\text{det}A} = \frac{10}{2} = 5,$$
    
$$x_2 = \frac{\text{det}A_2}{\text{det}A} = \frac{4}{2} = 2.$$

### Задание 2 <a class='anchor' id='slau2_2'>

Определим общую функцию для решения системы с треугольной матрицей

```python
def solve_triangle(A, b, reverse=False):
    """
    Решает систему уравнений для треугольной матрицы
    :param A: матрица коэффициентов
    :param b: правая часть
    :param reverse: решение для нижнетреугольной или верхнетреугольной матрицы
    :rtype: np.array
    """
    n_rows, n_cols = A.shape
    
    # Вектор решения     
    x = np.zeros(n_rows)
    indexes = range(n_rows) if not reverse else reversed(range(n_rows))
    for i in indexes:
        I = np.r_[0:i, i:n_cols]
        x[i] = (b[i] - x[I] @ A[i, I]) / A[i, i]
    return x
```

**Правило Крамера** <a class='anchor' id='slau2_5_1'>

```python
def cramer(A, b):
    """ 
    Решает систему равнений с помощью правила Крамера
    :param A: матрица коэффициентов системы
    :param b: вектор правой части
    :rtype: np.array
    """
    if A.shape[0] != A.shape[1]:
        raise Exception('Матрица не квадратная')
        
    # Определитель матрицы     
    det_A = np.linalg.det(A)
    if np.isclose(det_A, 0.0):
        raise Exception('Матрица вырождена')
    
    # Вектор решения     
    x = np.zeros(b.size)
    for col in range(A.shape[1]):
        A_i = np.copy(A)
        # Заменяем столбец матрицы A правой частью
        A_i[:, col] = b
        # Вычисляем значение неизвестной
        x[col] = np.linalg.det(A_i) / det_A
    return x
```

**LU-разложение** <a class='anchor' id='slau2_5_2'>

```python
def lu(A, b):
    """
    Решает систему уравнений с помощью LU-разложения
    :param A: матрица коэффициентов системы
    :param b: вектор правой части
    :rtype: np.array
    """
    n_rows, n_cols = A.shape
    
    if n_rows != n_cols:
        raise Exception('Матрица не квадратная')
        
    if np.isclose(np.linalg.det(A), 0.0):
        raise Exception('Матрица вырождена')
    
    # Матрицы LU-разложения     
    U = np.zeros(A.shape)
    L = np.eye(n_rows)
    
    # Вычислим матрицы LU-разложения     
    for i in range(n_rows):
        for j in range(n_cols):
            if i <= j:
                U[i, j] = A[i, j] - L[i, :i] @ U[:i, j]
            if i > j:
                L[i, j] = (A[i, j] - L[i, :j] @ U[:j, j]) / U[j, j]
    
    # Решаем систему Ly = b
    y = solve_triangle(L, b)
    # Решаем систему Ux = y
    x = solve_triangle(U, y, reverse=True)
    
    return x
```

**Метод Холецкого** <a class='anchor' id='slau2_5_3'>

```python
def is_symmetric(A):
    """
    Определяет, является ли матрица симметричной
    :param A: иссследуемая матрица
    :rtype: bool
    """
    return np.allclose(A, A.T)

def is_positive_defined(A):
    """
    Определяет, является ли матрица положительно определенной
    :param A: иссследуемая матрица
    :rtype: bool
    """
    return np.all(np.linalg.eigvals(A) > 0)


def cholesky(A, b):
    """
    Решает систему уравнений с помощью метода Холецкого
    :param A: матрица коэффициентов системы, невырожденная, симметричная и положительно определенная
    :param b: вектор правой части
    :rtype: np.array
    """
    if np.isclose(np.linalg.det(A), 0.0):
        raise Exception('Матрица вырождена')
        
    if not is_symmetric(A):
        raise Exception('Матрица не симметричная')
        
    if not is_positive_defined(A):
        raise Exception('Матрица не является положительно определенной')
    
    L = np.zeros(A.shape)
    # Алгоритм расчета матрицы L методом Холецкого     
    for i in range(A.shape[0]):
        for j in range(i + 1):
            L[i, j] = np.sqrt(A[i, j] - L[i, :i] @ L[i, :i]) if i == j else (A[i, j] - L[i, :j] @ L[j, :j]) / L[j, j]
    
    # Решаем систему Ly = b
    y = solve_triangle(L, b)
    # Решаем систему L^Tx = y
    x = solve_triangle(L.T, y, reverse=True)
    
    return x
```
